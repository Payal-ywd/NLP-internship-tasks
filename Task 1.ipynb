```python
import re
from collections import Counter

# Preprocessing function
def preprocess_text(text):
    if not text or not isinstance(text, str):
        return [], ""

    # Remove URLs & emails
    text = re.sub(r'http\S+|www\S+|https\S+|\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Convert to lowercase
    text = text.lower()

    # Remove repeated characters (soooo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Remove special characters & emojis
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenization
    tokens = text.split()

    # Remove short words except 'no', 'not'
    cleaned_tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]

    cleaned_sentence = " ".join(cleaned_tokens)

    return cleaned_tokens, cleaned_sentence
```

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for sentence in test_sentences:
    tokens, cleaned = preprocess_text(sentence)
    print("Original:", sentence)
    print("Tokens:", tokens)
    print("Cleaned:", cleaned)
    print("-" * 50)

    results.append(tokens)
```

```python
for i, tokens in enumerate(results):
    if tokens:
        total = len(tokens)
        unique = len(set(tokens))
        avg_len = sum(len(word) for word in tokens) / total
    else:
        total, unique, avg_len = 0, 0, 0

    print(f"Sentence {i+1}:")
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Avg Length:", round(avg_len, 2))
    print("-" * 40)
```

```python
all_tokens = [word for sublist in results for word in sublist]

counter = Counter(all_tokens)

print("Top 10 Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[:-6:-1])
```

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

# Test
pipeline_output = full_pipeline(test_sentences)
print(pipeline_output)
```

```python
edge_cases = ["", "😂😂😂😂", "123456789"]

for case in edge_cases:
    tokens, cleaned = preprocess_text(case)
    print("Input:", case)
    print("Tokens:", tokens)
    print("Cleaned:", cleaned)
    print("-" * 40)
```
